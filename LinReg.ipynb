```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress 
import statsmodels.api as sm
from statsmodels.sandbox.regression.predstd import wls_prediction_std
```

```python
df15 = pd.read_csv('UpdatedState2015.csv')
#adds column 'DeathsNorm', deaths per million in 2015
df15['DeathsNorm'] = (df15['Deaths']*(1e6)/df15['Population'])
df15state=df15[1:51]
df15state.at[5,'Mcdonalds']=1492
df15state.at[44,'Mcdonalds']=1224

df15state['Mcdonalds']=(pd.to_numeric(df15state['Mcdonalds'], errors='coerce'))
df15state
```

```python
# Regression line and results for populations (predictor variable) to death (response variable)
model = np.polyfit(df15state['Population'], df15state['Deaths'],1)
modelFit = np.poly1d(model)
plt.plot(df15state['Population'],df15state['Deaths'], 'yo',df15state['Population'], modelFit(df15state['Population']))
linregress(df15state['Population'], df15state['Deaths']) #x and y are arrays or lists.
plt.ylabel('Deaths') 
plt.xlabel('State Population')
plt.title("State Population and Deaths per State")

x = df15state['Population']
y = df15state['Deaths']
fig, ax = plt.subplots(figsize=(6, 4))
ax.scatter(x, y, alpha=0.5, color='orchid')
fig.suptitle('Example Scatter Plot')
fig.tight_layout(pad=2); 
ax.grid(True)
fig.savefig('filename1.png', dpi=125)

#print(x.min())
#print(x.max())
min = x.min()
max = x.max()

x = sm.add_constant(x)
#model: y ~ x + c
model = sm.OLS(y, x)
fitted = model.fit()
x_p = np.linspace(min, max, 51)

x_p2 = sm.add_constant(x_p)
#print(x_p2)
y_p = fitted.predict(x_p2)
#print(fitted.params)     # the estimated parameters for the regression line
#print(fitted.summary())  # summary statistics for the regression
ax.plot(x_p, y_p, '-', color='darkorchid', linewidth=2)

y_hat = fitted.predict(x) # x is an array from line 12 above
# print(y_hat)
y_err = y - y_hat
mean_x = x.T[1].mean()
n = len(x) # length of population vector
dof = n - fitted.df_model - 1 # degrees of freedom
from scipy import stats
t = stats.t.ppf(1-0.025, df=dof) # for 2 tailed, use this t val of alpha/2
s_err = np.sum(np.power(y_err, 2))
conf = t * np.sqrt((s_err/(n-2))*(1.0/n + (np.power((x_p-mean_x),2) / 
       ((np.sum(np.power(x_p,2))) - n*(np.power(mean_x,2))))))
upper = y_p + abs(conf)
lower = y_p - abs(conf)
ax.fill_between(x_p,lower,upper,color='#888888', alpha=0.4)
fig.savefig('filename3.png', dpi=125)
sdev, lowerPI, upperPI = wls_prediction_std(fitted, exog=x_p2, alpha=0.05)
ax.fill_between(x_p, lowerPI, upperPI, color='#888888', alpha=0.1)
fig.savefig('filename4.png', dpi=125)
```

```python
# Regression line and results for Gini (predictor variable) to Death/1M (response variable)
x='Poverty'
y='DeathsNorm'

model = np.polyfit(df15state[x], df15state[y],1)
modelFit = np.poly1d(model)

### testing plotting CIs
cf_fit = modelFit(df15state[x])
# get the coordinates for the fit curve
c_y = [np.min(cf_fit),np.max(cf_fit)]
c_x = [np.min(df15state[x]),np.max(df15state[x])]
# predict y values of original data using the fit (lin regress)
p_y = model[0] * df15state[x] + model[1]
# calculate the y-error (residuals)
y_err = df15state[y] - p_y
# create series of new test x-values to predict for
p_x = np.arange(np.min(df15state[x]),np.max(df15state[x])+1,1)

# now calculate confidence intervals for new test x-series
mean_x = np.mean(df15state[x])         # mean of x
n = len(df15state[x])              # number of samples in original fit
t = 2.011               # appropriate t value (where n=50, two tailed 95%)
s_err = np.sum(np.power(y_err,2))   # sum of the squares of the residuals
#print(s_err)
#print(y_err)
#print(mean_x)

# calculate your CIs
confs = t * np.sqrt((s_err/(n-2))*(1.0/n + (np.power((p_x-mean_x),2)/
            ((np.sum(np.power(df15state[x],2)))-n*(np.power(mean_x,2))))))
# now predict y based on test x-values
p_y = model[0]*p_x+model[1]
#print(p_y)
 
# get lower and upper confidence limits based on predicted y and confidence intervals
lower = p_y - abs(confs)
upper = p_y + abs(confs)

plt.axes().set_aspect('equal')
plt.plot(df15state[x],df15state[y], 'yo',df15state[x], modelFit(df15state[x]))
linres=linregress(df15state[x], df15state[y]) #x and y are arrays or lists.
plt.ylabel(y) 
plt.xlabel(x)

### for CIs
# plot confidence limits
plt.plot(p_x,lower,'b--',label='Lower confidence limit (95%)')
plt.plot(p_x,upper,'b--',label='Upper confidence limit (95%)')

# now plot the PIs
# from statsmodels.sandbox.regression.predstd import wls_prediction_std
# import statsmodels.api as sm

# x = sm.add_constant(df15state[x])
# model = sm.OLS(df15state[y], df15state[x])
# fitted = model.fit()
# x_pred = np.linspace(df15state[x].min(), df15state[x].max(), 48)
# x_const = sm.add_constant(df15state[x])
# x_pred2 = sm.add_constant(x_pred)

# sdev, lowerPI, upperPI = wls_prediction_std(fitted, exog=x_pred2, alpha=0.05)
# print(p_x)

plt.title("Deaths Normalized per Poverty Level")

linres
```
